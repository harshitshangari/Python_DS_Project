```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
data = pd.read_parquet(Path("data") / "train.parquet")
```

```python
df1 = pd.read_csv('2020.csv')
```

```python
df2 = pd.read_csv('2021.csv')
```

```python
weather = pd.concat([df1, df2], ignore_index=True)
```

```python
weather.head(10)
```

```python
weather = weather.drop(columns=['snow', 'wpgt','tsun','coco','Coco'])
```

```python
from datetime import datetime, timedelta
#weather['date'] = datetime.strptime(weather['date'], '%Y-%m-%d %H:%M:%S')
weather["date"] = pd.to_datetime(weather["date"])
weather["hour"] = pd.to_timedelta(weather['hour'], unit='h')
weather['date'] = weather['date'] + weather['hour']
weather = weather.drop(columns=['hour'])
```

```python
full_data = data.merge(weather, on='date')
full_data
```

```python
#CAN BE DONE USING SIMPLEIMPUTER IN PIPELINING
full_data = full_data.fillna(method='bfill')
full_data
```

```python
def _encode_dates(X):
    X = X.copy()  # modify a copy of X
    # Encode the date information from the DateOfDeparture columns
    X.loc[:, "year"] = X["date"].dt.year
    X.loc[:, "month"] = X["date"].dt.month
    X.loc[:, "day"] = X["date"].dt.day
    X.loc[:, "weekday"] = X["date"].dt.weekday
    X.loc[:, "hour"] = X["date"].dt.hour

    # Finally we can drop the original columns from the dataframe
    return X.drop(columns=["date"])
```

```python
from sklearn.preprocessing import FunctionTransformer

date_encoder = FunctionTransformer(_encode_dates, validate=False)
date_encoder.fit_transform(full_data[["date"]]).head()
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse=False)

enc.fit_transform(_encode_dates(full_data[["date"]])[["hour"]])
```

```python
import problem

X_train, y_train = problem.get_train_data()
X_test, y_test = problem.get_test_data()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline

date_encoder = FunctionTransformer(_encode_dates)
date_cols = _encode_dates(X_train[["date"]]).columns.tolist()

categorical_encoder = OneHotEncoder(handle_unknown="ignore")
categorical_cols = ["counter_name", "site_name"]

preprocessor = ColumnTransformer(
    [
        ("date", OneHotEncoder(handle_unknown="ignore"), date_cols),
        ("cat", categorical_encoder, categorical_cols),
    ]
)

regressor = Ridge()

pipe = make_pipeline(date_encoder, preprocessor, regressor)
X_train.head()
#pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

print(
    f"Train set, RMSE={mean_squared_error(y_train, pipe.predict(X_train), squared=False):.2f}"
)
print(
    f"Test set, RMSE={mean_squared_error(y_test, pipe.predict(X_test), squared=False):.2f}"
)
```
